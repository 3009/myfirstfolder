```python
import numpy as np
import matplotlib.pyplot as plt
angles=np.arange(0,360)
sin=np.linspace(1,2*np.pi,360)
cos=np.linspace(1,2*np.pi,360)
sin=np.sin(sin)
cos=np.cos(cos)
```

```python
plt.plot(angles,sin,'r:')
plt.plot(angles,cos,'g:')
plt.show()
```
